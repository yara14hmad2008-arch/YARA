```python
df = pd.read_excel("/content/2قاعدة البيانات النهائية 190 سؤال(بعد).xlsx")  # قراءة ملف قاعدة البيانات الذي يحتوي على الاسئلة والاجوبة

questions = df["السؤال"].tolist()  # استخراج عمود الاسئلة وتحويله الى قائمة لاستخدامه في عملية المطابقة

answers = df["الإجابة"].tolist()  # استخراج عمود الاجابات وتحويله الى قائمة ليتم ارجاع الاجابة المناسبة عند العثور على تطابق

len(df)  # عرض عدد الاسئلة الموجودة في قاعدة البيانات للتأكد من تحميل جميع البيانات بنجاح
```

```python
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")#نموذج الفهم الذكي عشان يخلي الروبوت يفهم جميع الاسئلة حتى لو ما كانت حرفي يعني حتى لو قريب منها
```

```python
question_embeddings = model.encode(questions, convert_to_tensor=True)# تحويل الاسئلة من نص الى تمثيل عددي حتى يستطيع النموذج فهم المعنى ومقارنة تشابه الاسئلة
```

```python
def chatbot(user_question):# دالة تستقبل سؤال الطالب وتحسب اقرب سؤال في قاعدة البيانات بناء على التشابه ثم ترجع الجواب المناسب
    user_embedding = model.encode(user_question, convert_to_tensor=True)
    similarities = util.cos_sim(user_embedding, question_embeddings)
    best_match_idx = torch.argmax(similarities)
    return answers[best_match_idx]
```

```python
# استيراد مكتبة ipywidgets لإنشاء واجهات تفاعلية
import ipywidgets as widgets
# استيراد display و clear_output لعرض العناصر وتحديثها
from IPython.display import display, clear_output

# قاموس لتخزين بيانات الطالب
student_data = {}

# قائمة لتخزين رسائل المحادثة
chat_history = []

# صندوق يحتوي حقول إدخال البيانات
form_box = widgets.VBox()

# عرض صندوق الإدخال
display(form_box)

# ========================
# مرحلة إدخال الاسم
# ========================
def show_name():
    form_box.children = []  # مسح أي عنصر سابق

    # إنشاء حقل إدخال الاسم
    name_input = widgets.Text(
        placeholder="😀ما اسمك؟",
        layout=widgets.Layout(width='400px')
    )

    # عنصر لعرض رسالة الخطأ
    error = widgets.HTML(value="", layout=widgets.Layout(height='20px'))

    # دالة حفظ الاسم عند الضغط Enter
    def save_name(change):
        name = name_input.value.strip()  # إزالة المسافات

        # التحقق أن الاسم لا يحتوي أرقام
        if not name.isalpha():
            error.value = "<span style='color:red'>😊الرجاء ادخال اسم بدون ارقام</span>"
        else:
            student_data["name"] = name  # تخزين الاسم
            show_age()  # الانتقال لمرحلة العمر

    # تنفيذ الدالة عند الضغط Enter
    name_input.on_submit(save_name)

    # عرض الحقل ورسالة الخطأ
    form_box.children = [name_input, error]


# ========================
# مرحلة إدخال العمر
# ========================
def show_age():
    form_box.children = []

    age_input = widgets.Text(
        placeholder="😉كم عمرك؟",
        layout=widgets.Layout(width='400px')
    )

    error = widgets.HTML(value="", layout=widgets.Layout(height='20px'))

    def save_age(change):
        age = age_input.value.strip()

        # التحقق أن العمر أرقام فقط
        if not age.isdigit():
            error.value = "<span style='color:red'>😊الرجاء ادخال العمر بالارقام فقط</span>"
        else:
            student_data["age"] = age
            show_grade()  # الانتقال لمرحلة الصف

    age_input.on_submit(save_age)
    form_box.children = [age_input, error]


# ========================
# مرحلة إدخال الصف (بدون تحقق)
# ========================
def show_grade():
    form_box.children = []

    grade_input = widgets.Text(
        placeholder="👀ما صفك الدراسي؟",
        layout=widgets.Layout(width='400px')
    )

    def save_grade(change):
        student_data["grade"] = grade_input.value.strip()
        show_school()  # الانتقال لمرحلة المدرسة

    grade_input.on_submit(save_grade)
    form_box.children = [grade_input]


# ========================
# مرحلة إدخال المدرسة
# ========================
def show_school():
    form_box.children = []

    school_input = widgets.Text(
        placeholder="🏫من اي مدرسة انت؟",
        layout=widgets.Layout(width='400px')
    )

    error = widgets.HTML(value="", layout=widgets.Layout(height='20px'))

    def save_school(change):
        school = school_input.value.strip()

        # السماح بالحروف والمسافات فقط
        if not school.replace(" ", "").isalpha():
            error.value = "<span style='color:red'>😊الرجاء ادخال اسم مدرسة صحيح بدون ارقام</span>"
        else:
            student_data["school"] = school
            form_box.children = []  # إخفاء الفورم
            start_chat()  # بدء المحادثة

    school_input.on_submit(save_school)
    form_box.children = [school_input, error]


# ========================
# إنشاء واجهة الشات
# ========================

# صندوق عرض المحادثة
chat_box = widgets.HTML(
    layout=widgets.Layout(
        height='500px',
        overflow_y='auto',
        border='none',
        padding='15px'
    )
)

# حقل إدخال الرسائل
user_input = widgets.Text(
    placeholder="😄اكتب سؤالك هنا...",
    layout=widgets.Layout(width='85%', height='45px')
)

# زر الإرسال
send_button = widgets.Button(
    description="ارسال",
    layout=widgets.Layout(width='15%', height='45px')
)

# تجميع الحقل والزر بجانب بعض
input_row = widgets.HBox([user_input, send_button])


# ========================
# دالة عرض المحادثة
# ========================
def render_chat():
    # إنشاء تصميم الشات باستخدام HTML
    html = """
    <div style='direction:rtl; font-family:Cairo, Arial; background:#1e1e1e; padding:15px; border-radius:15px;'>
    """

    # المرور على كل رسالة
    for sender, message in chat_history:
        if sender == "user":
            label = student_data["name"]
            align = "right"
            bubble_color = "linear-gradient(135deg, #4e73df, #224abe)"
        else:
            label = "🤖الروبوت"
            align = "left"
            bubble_color = "linear-gradient(135deg, #2d2d2d, #3a3a3a)"

        # إنشاء فقاعة الرسالة
        html += f"""
        <div style="text-align:{align}; margin:12px;">
            <div style="font-size:11px; color:#aaa; margin-bottom:4px;">{label}</div>
            <div style="
                background:{bubble_color};
                color:white;
                padding:12px 18px;
                border-radius:22px;
                display:inline-block;
                max-width:70%;
                box-shadow:0 4px 10px rgba(0,0,0,0.3);
                font-size:14px;
            ">
            {message}
            </div>
        </div>
        """

    html += "</div>"
    chat_box.value = html  # تحديث عرض الشات


# ========================
# دالة معالجة الرسائل
# ========================
def process_message(b=None):
    message = user_input.value.strip()

    # منع إرسال رسالة فارغة
    if message == "":
        return

    chat_history.append(("user", message))  # إضافة رسالة المستخدم

    # إذا كتب خروج
    if message == "📤خروج":
        chat_history.append(("bot", "بالتوفيق 👋"))
    else:
        answer = chatbot(message)  # استدعاء دالة الروبوت
        chat_history.append(("bot", answer))

    user_input.value = ""  # تفريغ الحقل
    render_chat()  # تحديث الشات


# ربط زر الإرسال
send_button.on_click(process_message)

# ربط الإرسال بزر Enter
user_input.on_submit(process_message)


# ========================
# بدء الشات
# ========================
def start_chat():
    display(widgets.HTML(f"<h3 style='direction:rtl'>اهلا {student_data['name']} 👋</h3>"))
    display(chat_box)
    display(input_row)
    chat_history.append(("bot", "✨ جاهزين نبدأ ✨"))
    render_chat()


# بدء أول خطوة (إظهار حقل الاسم)
show_name()
```
