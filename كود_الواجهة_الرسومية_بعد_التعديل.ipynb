```python
df = pd.read_excel("/content/2قاعدة البيانات النهائية 190 سؤال(بعد).xlsx")  # قراءة ملف قاعدة البيانات الذي يحتوي على الاسئلة والاجوبة

questions = df["السؤال"].tolist()  # استخراج عمود الاسئلة وتحويله الى قائمة لاستخدامه في عملية المطابقة

answers = df["الإجابة"].tolist()  # استخراج عمود الاجابات وتحويله الى قائمة ليتم ارجاع الاجابة المناسبة عند العثور على تطابق

len(df)  # عرض عدد الاسئلة الموجودة في قاعدة البيانات للتأكد من تحميل جميع البيانات بنجاح
```

```python
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")#نموذج الفهم الذكي عشان يخلي الروبوت يفهم جميع الاسئلة حتى لو ما كانت حرفي يعني حتى لو قريب منها
```

```python
question_embeddings = model.encode(questions, convert_to_tensor=True)# تحويل الاسئلة من نص الى تمثيل عددي حتى يستطيع النموذج فهم المعنى ومقارنة تشابه الاسئلة
```

```python
def chatbot(user_question):
    # هنا أنشأت دالة اسمها chatbot
    # تستقبل سؤال المستخدم كنص داخل المتغير user_question
    # هدف الدالة هو إيجاد أقرب سؤال مشابه في قاعدة البيانات وإرجاع جوابه

    user_embedding = model.encode(user_question, convert_to_tensor=True)
    # في هذا السطر قمت بتحويل سؤال المستخدم من نص عادي
    # إلى تمثيل عددي يسمى embedding
    # هذا التمثيل العددي يساعد النموذج على فهم معنى السؤال وليس فقط كلماته
    # استخدمت convert_to_tensor=True حتى أحصل على الناتج بصيغة Tensor
    # لأن حساب التشابه يحتاج إلى Tensor

    similarities = util.cos_sim(user_embedding, question_embeddings)
    # هنا قمت بحساب درجة التشابه بين سؤال المستخدم
    # وبين جميع الأسئلة الموجودة في قاعدة البيانات
    # استخدمت cosine similarity لأنها تقيس مدى تقارب المعاني بين الجمل
    # الناتج يكون مصفوفة تحتوي على درجة تشابه لكل سؤال

    best_match_idx = torch.argmax(similarities)
    # في هذا السطر حددت رقم السؤال الذي حصل على أعلى درجة تشابه
    # argmax ترجع موقع أعلى قيمة داخل مصفوفة التشابه
    # هذا يعني أنني اخترت السؤال الأقرب في المعنى

    return answers[best_match_idx]
    # أخيراً أرجعت الجواب الموجود في نفس موقع السؤال الأقرب
    # وبالتالي يحصل المستخدم على الجواب المناسب بناءً على أقرب تطابق معنوي
```

```python
# قمت باستيراد مكتبة ipywidgets لاستخدام عناصر واجهة المستخدم
# مثل مربعات النص والأزرار داخل بيئة Google Colab أو Jupyter Notebook
import ipywidgets as widgets

# هنا قمت باستيراد الدالة display من مكتبة IPython.display
# وذلك حتى أتمكن من عرض عناصر الواجهة التي أنشئها على الشاشة
from IPython.display import display

# أنشأت قاموس فارغ باسم student_data
# هذا القاموس سأستخدمه لاحقًا لتخزين بيانات الطالب
# مثل الاسم أو الرقم أو أي معلومات أخرى يتم إدخالها
student_data = {}

# أنشأت قائمة فارغة باسم chat_history
# هذه القائمة سأستخدمها لتخزين سجل المحادثة
# أي كل سؤال وجواب يتم بين المستخدم والروبوت سيتم حفظه هنا
chat_history = []

# ======================
# تنسيق عام بنفسجي متناسق
# ======================
# هنا استخدمت الدالة display لعرض كود HTML داخل بيئة العمل
# الهدف من ذلك هو تطبيق تنسيقات CSS مخصصة على واجهة التطبيق
display(widgets.HTML("""

<style>

/* قمت بتنسيق خلفية الصفحة كاملة */
body {
    /* أضفت تدرج لوني عمودي ليعطي مظهرًا داكنًا واحترافيًا */
    background: linear-gradient(180deg,#1e1a2e,#26213d);
}

/* هذا التنسيق خاص بالعنوان الرئيسي في الواجهة */
.title-main {
    text-align:center;        /* جعلت العنوان في المنتصف */
    color:#e6dff3;            /* اخترت لونًا فاتحًا ليتناسب مع الخلفية الداكنة */
    font-size:30px;           /* حددت حجم الخط ليكون واضحًا وكبيرًا */
    font-weight:bold;         /* جعلت الخط عريضًا ليبرز أكثر */
    margin-bottom:25px;       /* أضفت مسافة أسفل العنوان للفصل بينه وبين العناصر التالية */
}

/* هذا التنسيق يمثل بطاقة النموذج التي تحتوي على الحقول */
.form-card {
    background:#2f2946;       /* لون خلفية البطاقة */
    padding:25px;             /* مسافة داخلية لإعطاء شكل منظم */
    border-radius:14px;       /* تدوير الحواف ليصبح الشكل أكثر حداثة */
    width:75%;                /* جعلت عرض البطاقة 75 بالمئة من الصفحة */
    margin:auto;              /* وضعتها في المنتصف أفقيًا */
    box-shadow:0 8px 20px rgba(0,0,0,0.45); /* أضفت ظلًا لإعطاء تأثير ثلاثي الأبعاد */
}

/* تنسيق حقول الإدخال */
.input-dark input {
    background:#3a3356 !important; /* لون خلفية الحقل */
    color:white !important;        /* لون النص داخل الحقل */
    border:1px solid #B2A1C7 !important; /* لون الإطار */
}

/* تنسيق الأزرار */
.dark-button button {
    background:#B2A1C7 !important;  /* لون خلفية الزر */
    color:#1e1a2e !important;       /* لون النص داخل الزر */
    width:100%;                    /* جعلت الزر بعرض الحاوية */
    height:45px;                   /* تحديد ارتفاع الزر */
    font-size:16px;                /* حجم الخط */
    font-weight:bold;              /* جعل النص عريضًا */
    border-radius:8px !important;  /* تدوير حواف الزر */
}

/* تنسيق حاوية الدردشة */
.chat-container {
    background:#2f2946;  /* نفس لون البطاقة للحفاظ على التناسق */
    padding:25px;        /* مسافة داخلية */
    border-radius:14px;  /* تدوير الحواف */
}

</style>

"""))

form_box = widgets.VBox()
display(form_box)

# ======================
# الاسم
# ======================
# قمت بتعريف دالة باسم show_name
# هذه الدالة مسؤولة عن عرض واجهة إدخال اسم الطالب
def show_name():

    # في البداية قمت بتفريغ محتوى form_box
    # وذلك حتى يتم حذف أي عناصر سابقة وعرض الواجهة الجديدة فقط
    form_box.children = []

    # أنشأت عنوانًا رئيسيًا باستخدام HTML
    # وأضفت له الفئة title-main حتى يأخذ التنسيق الذي صممته سابقًا في CSS
    title = widgets.HTML("<div class='title-main'>✨روبوت الاسئلة التعليمية✨</div>")

    # أنشأت حقل إدخال نصي لاستقبال اسم الطالب
    # وحددت عرض الحقل ليكون 100 بالمئة من عرض البطاقة
    name_input = widgets.Text(description="😀ما اسمك؟", layout=widgets.Layout(width='100%'))

    # أضفت له فئة input-dark حتى يأخذ التصميم الداكن الذي أنشأته في التنسيقات
    name_input.add_class("input-dark")

    # أنشأت عنصر HTML لعرض رسائل الخطأ
    # وحددت ارتفاعه حتى يبقى مكان الرسالة ثابتًا في التصميم
    error = widgets.HTML(layout=widgets.Layout(height='20px'))

    # أنشأت زرًا باسم إرسال
    send_btn = widgets.Button(description="✒️إرسال")

    # أضفت له فئة dark-button ليأخذ تنسيق الأزرار المخصص
    send_btn.add_class("dark-button")

    # عرفت دالة داخلية باسم save_name
    # هذه الدالة مسؤولة عن حفظ الاسم بعد التحقق منه
    def save_name(b=None):

        # أخذت القيمة المدخلة من الحقل وحذفت الفراغات الزائدة
        name = name_input.value.strip()

        # تحققت أن الاسم يحتوي على أحرف فقط بدون أرقام
        if not name.isalpha():

            # في حال كان الإدخال غير صحيح أعرض رسالة خطأ للمستخدم
            error.value = "<span style='color:#ffb4a2'>😊الرجاء ادخال اسم بدون ارقام</span>"

        else:
            # إذا كان الاسم صحيحًا أقوم بتخزينه داخل قاموس student_data
            student_data["name"] = name

            # ثم أنتقل إلى المرحلة التالية وهي عرض إدخال العمر
            show_age()

    # ربطت زر الإدخال عند الضغط على زر Enter بتنفيذ دالة الحفظ
    name_input.on_submit(save_name)

    # وربطت أيضًا زر الإرسال بتنفيذ نفس الدالة عند الضغط عليه
    send_btn.on_click(save_name)

    # جمعت جميع العناصر داخل VBox حتى تظهر بشكل عمودي مرتب
    card = widgets.VBox([title, name_input, error, send_btn])

    # أضفت فئة form-card حتى يأخذ التصميم الخاص بالبطاقة
    card.add_class("form-card")

    # أخيرًا قمت بعرض البطاقة داخل form_box
    form_box.children = [card]

# ======================
# العمر
# ======================

# قمت بإنشاء دالة باسم show_age
# هذه الدالة مسؤولة عن عرض نموذج إدخال العمر للمستخدم
def show_age():

    # هنا أقوم بتفريغ محتويات form_box
    # الهدف هو إزالة أي عناصر سابقة قبل عرض نموذج العمر
    form_box.children = []

    # أنشأت مربع إدخال نصي باستخدام widgets.Text
    # هذا الحقل مخصص لإدخال عمر الطالب
    # جعلت العرض 100 بالمئة ليأخذ كامل عرض البطاقة
    age_input = widgets.Text(description="😉كم عمرك؟", layout=widgets.Layout(width='100%'))

    # أضفت له كلاس تنسيقي باسم input-dark
    # وذلك ليتم تطبيق تنسيق CSS الداكن عليه
    age_input.add_class("input-dark")

    # أنشأت عنصر HTML لعرض رسائل الخطأ
    # وحددت ارتفاعه حتى لا يتغير شكل الواجهة عند ظهور الرسالة
    error = widgets.HTML(layout=widgets.Layout(height='20px'))

    # أنشأت زر باسم إرسال
    send_btn = widgets.Button(description="✒️إرسال")

    # أضفت له كلاس dark-button لتطبيق تنسيق الزر المخصص
    send_btn.add_class("dark-button")

    # أنشأت دالة داخلية باسم save_age
    # هذه الدالة يتم تنفيذها عند الضغط على الزر أو عند الضغط على Enter
    def save_age(b=None):

        # هنا أقوم بأخذ القيمة المدخلة وإزالة الفراغات الزائدة
        age = age_input.value.strip()

        # أتحقق إذا كانت القيمة المدخلة أرقام فقط
        # في حال لم تكن أرقام يتم عرض رسالة خطأ
        if not age.isdigit():
            error.value = "<span style='color:#ffb4a2'>😊الرجاء ادخال العمر بالارقام فقط</span>"

        # إذا كانت القيمة صحيحة يتم تخزينها داخل قاموس student_data
        # ثم الانتقال للمرحلة التالية وهي اختيار الصف
        else:
            student_data["age"] = age
            show_grade()

    # هنا أربط الضغط على Enter داخل حقل الإدخال بالدالة save_age
    age_input.on_submit(save_age)

    # هنا أربط الضغط على زر إرسال بنفس الدالة
    send_btn.on_click(save_age)

    # أنشأت بطاقة تحتوي على العناصر الثلاثة
    # مربع الإدخال ورسالة الخطأ والزر
    card = widgets.VBox([age_input, error, send_btn])

    # أضفت كلاس form-card لتطبيق تصميم البطاقة
    card.add_class("form-card")

    # أخيرًا أقوم بعرض البطاقة داخل form_box
    form_box.children = [card]

# ======================
# الصف
# ======================

# قمت بإنشاء دالة باسم show_grade
# هذه الدالة مسؤولة عن عرض نموذج إدخال الصف بعد إدخال العمر
def show_grade():

    # هنا أقوم بتفريغ محتويات form_box
    # الهدف هو إزالة نموذج العمر واستبداله بنموذج الصف
    form_box.children = []

    # أنشأت مربع إدخال نصي خاص بإدخال الصف الدراسي
    # حددت العرض 100 بالمئة ليأخذ كامل عرض البطاقة
    grade_input = widgets.Text(description="👀ما صفك الدراسي؟", layout=widgets.Layout(width='100%'))

    # أضفت كلاس input-dark لتطبيق التنسيق الداكن على حقل الإدخال
    grade_input.add_class("input-dark")

    # أنشأت زر إرسال ليقوم المستخدم بتأكيد إدخال الصف
    send_btn = widgets.Button(description="✒️إرسال")

    # أضفت له كلاس dark-button لتطبيق تصميم الأزرار المخصص
    send_btn.add_class("dark-button")

    # أنشأت دالة داخلية باسم save_grade
    # هذه الدالة يتم تنفيذها عند الضغط على Enter أو على زر الإرسال
    def save_grade(b=None):

        # أقوم بأخذ القيمة المدخلة وإزالة الفراغات الزائدة من البداية والنهاية
        # ثم أحفظها داخل قاموس student_data تحت المفتاح grade
        student_data["grade"] = grade_input.value.strip()

        # بعد حفظ الصف أنتقل مباشرة إلى المرحلة التالية وهي إدخال المدرسة
        show_school()

    # أربط الضغط على Enter داخل مربع الإدخال بالدالة save_grade
    grade_input.on_submit(save_grade)

    # أربط الضغط على زر إرسال بنفس الدالة
    send_btn.on_click(save_grade)

    # أنشأت بطاقة تحتوي على مربع الإدخال والزر
    card = widgets.VBox([grade_input, send_btn])

    # أضفت كلاس form-card لتطبيق تنسيق البطاقة على هذا النموذج
    card.add_class("form-card")

    # أخيرًا أعرض البطاقة داخل form_box ليظهر نموذج الصف للمستخدم
    form_box.children = [card]


# ======================
# المدرسة
# ======================

# قمت بإنشاء دالة باسم show_school
# هذه الدالة مسؤولة عن عرض نموذج إدخال اسم المدرسة
# وهي المرحلة الأخيرة من جمع بيانات الطالب قبل بدء الدردشة
def show_school():

    # هنا أقوم بتفريغ محتويات form_box
    # وذلك لإزالة نموذج الصف واستبداله بنموذج المدرسة
    form_box.children = []

    # أنشأت مربع إدخال نصي لإدخال اسم المدرسة
    # حددت العرض 100 بالمئة ليأخذ كامل عرض البطاقة
    school_input = widgets.Text(description="🏫من اي مدرسة انت؟", layout=widgets.Layout(width='100%'))

    # أضفت كلاس input-dark لتطبيق التنسيق الداكن على حقل الإدخال
    school_input.add_class("input-dark")

    # أنشأت عنصر HTML لعرض رسائل الخطأ عند إدخال بيانات غير صحيحة
    # وحددت ارتفاعه للحفاظ على ثبات شكل الواجهة
    error = widgets.HTML(layout=widgets.Layout(height='20px'))

    # أنشأت زر إرسال لتأكيد إدخال اسم المدرسة
    send_btn = widgets.Button(description="✒️إرسال")

    # أضفت له كلاس dark-button لتطبيق تنسيق الأزرار المخصص
    send_btn.add_class("dark-button")

    # أنشأت دالة داخلية باسم save_school
    # يتم تنفيذها عند الضغط على Enter أو على زر الإرسال
    def save_school(b=None):

        # أقوم بأخذ النص المدخل وإزالة الفراغات الزائدة من البداية والنهاية
        school = school_input.value.strip()

        # أتحقق من أن اسم المدرسة يحتوي على حروف فقط
        # أستخدم replace لإزالة المسافات ثم أتحقق باستخدام isalpha
        # إذا لم يكن الإدخال صحيحًا يتم عرض رسالة خطأ
        if not school.replace(" ", "").isalpha():
            error.value = "<span style='color:#ffb4a2'>😊الرجاء ادخال اسم مدرسة صحيح بدون ارقام</span>"

        # إذا كان الإدخال صحيحًا يتم حفظه داخل قاموس student_data
        # ثم يتم إفراغ النموذج والانتقال لبدء نظام الدردشة
        else:
            student_data["school"] = school
            form_box.children = []
            start_chat()

    # أربط الضغط على Enter داخل مربع الإدخال بالدالة save_school
    school_input.on_submit(save_school)

    # أربط الضغط على زر إرسال بنفس الدالة
    send_btn.on_click(save_school)

    # أنشأت بطاقة تحتوي على مربع الإدخال ورسالة الخطأ وزر الإرسال
    card = widgets.VBox([school_input, error, send_btn])

    # أضفت كلاس form-card لتطبيق تنسيق البطاقة الموحد على هذا النموذج
    card.add_class("form-card")

    # أخيرًا أعرض البطاقة داخل form_box ليظهر نموذج إدخال المدرسة للمستخدم
    form_box.children = [card]


# ======================
# واجهة الشات
# ======================

# هنا أنشأت صندوق HTML لعرض المحادثة
# حددت ارتفاعه 500 بكسل وجعلت التمرير العمودي مفعلاً عند زيادة الرسائل
chat_box = widgets.HTML(layout=widgets.Layout(height='500px', overflow_y='auto'))

# أضفت كلاس chat-container لتطبيق تنسيق CSS الخاص بحاوية الدردشة
chat_box.add_class("chat-container")

# أنشأت مربع إدخال نصي ليكتب المستخدم سؤاله
# أضفت نصًا إرشاديًا داخل الحقل
# وحددت العرض 85 بالمئة ليكون بجانب زر الإرسال
# وحددت الارتفاع ليكون متناسقًا مع الزر
user_input = widgets.Text(
    placeholder="😄اكتب سؤالك هنا...",
    layout=widgets.Layout(width='85%', height='45px')
)

# أضفت كلاس input-dark لتطبيق التصميم الداكن على حقل الإدخال
user_input.add_class("input-dark")

# أنشأت زر إرسال بجانب مربع الإدخال
# جعلت عرضه 15 بالمئة ليكمل المساحة المتبقية بجانب الحقل
# وحددت نفس الارتفاع للحفاظ على تناسق الشكل
send_button = widgets.Button(description="✒️إرسال", layout=widgets.Layout(width='15%', height='45px'))

# أضفت كلاس dark-button لتطبيق تنسيق الأزرار
send_button.add_class("dark-button")

# استخدمت HBox لوضع مربع الإدخال وزر الإرسال في صف أفقي واحد
input_row = widgets.HBox([user_input, send_button])

# أنشأت دالة باسم render_chat
# هذه الدالة مسؤولة عن إعادة رسم المحادثة بالكامل عند إضافة أي رسالة جديدة
def render_chat():

    # بدأت ببناء كود HTML مع تحديد اتجاه النص من اليمين إلى اليسار
    # وذلك لأن الواجهة باللغة العربية
    html = "<div style='direction:rtl;'>"

    # أتنقل عبر جميع الرسائل المخزنة في chat_history
    # كل عنصر عبارة عن مرسل ورسالة
    for sender, message in chat_history:

        # إذا كان المرسل هو المستخدم
        if sender == "user":
            align = "right"                 # أجعل الرسالة بمحاذاة اليمين
            color = "#8f7eb3"               # أحدد لون فقاعة المستخدم
            label = student_data["name"]   # أستخدم اسم الطالب كعنوان للرسالة

        # إذا كان المرسل هو الروبوت
        else:
            align = "left"                  # أجعل الرسالة بمحاذاة اليسار
            color = "#6d6487"               # لون مختلف لفقاعة الروبوت
            label = "Chatbot"               # اسم المرسل هو Chatbot

        # هنا أقوم بإضافة تنسيق HTML لكل رسالة
        # أضع اسم المرسل أعلى الفقاعة
        # ثم أضع نص الرسالة داخل فقاعة بتصميم محدد
        html += f"""
        <div style="text-align:{align}; margin:14px 0;">
            <div style="font-size:13px; color:#e6ddff;">{label}</div>
            <div style="background:{color}; color:white; padding:15px 20px; border-radius:20px; display:inline-block; max-width:70%; font-size:17px;">
            {message}
            </div>
        </div>
        """

    # إغلاق عنصر الاتجاه الرئيسي
    html += "</div>"

    # تحديث محتوى chat_box ليعرض المحادثة بعد إعادة بنائها
    chat_box.value = html

# هذه الدالة مسؤولة عن معالجة الرسالة عند ضغط زر الإرسال أو الضغط على Enter
def process_message(b=None):

    # أقوم بأخذ النص الذي كتبه المستخدم مع إزالة الفراغات الزائدة
    message = user_input.value.strip()

    # إذا كانت الرسالة فارغة لا يتم تنفيذ أي شيء
    if message == "":
        return

    # أضيف رسالة المستخدم إلى سجل المحادثة مع تحديد أن المرسل هو user
    chat_history.append(("user", message))

    # أتحقق إذا كانت الرسالة تساوي كلمة الخروج المحددة
    if message == "📤خروج":

        # إذا طلب المستخدم الخروج أضيف رسالة وداع من الروبوت
        chat_history.append(("bot", "😚بالتوفيق"))

    else:
        # إذا لم تكن رسالة خروج أقوم بإرسال السؤال إلى دالة chatbot
        # هذه الدالة تعالج السؤال وتعيد الإجابة المناسبة
        answer = chatbot(message)

        # أضيف إجابة الروبوت إلى سجل المحادثة
        chat_history.append(("bot", answer))

    # بعد معالجة الرسالة أقوم بتفريغ مربع الإدخال ليكون جاهزًا لرسالة جديدة
    user_input.value = ""

    # أعيد رسم المحادثة لتحديث العرض وإظهار الرسائل الجديدة
    render_chat()


# أربط زر الإرسال بدالة process_message
# بحيث يتم تنفيذها عند الضغط على الزر
send_button.on_click(process_message)

# أربط الضغط على Enter داخل مربع الإدخال بنفس الدالة
user_input.on_submit(process_message)


# هذه الدالة مسؤولة عن بدء واجهة الدردشة بعد إدخال بيانات الطالب
def start_chat():

    # أعرض عنوان الواجهة باستخدام تنسيق HTML
    display(widgets.HTML("<div class='title-main'>✨روبوت الاسئلة التعليمية✨</div>"))

    # أعرض صندوق المحادثة
    display(chat_box)

    # أعرض صف الإدخال الذي يحتوي على مربع النص وزر الإرسال
    display(input_row)

    # أحصل على اسم الطالب من قاموس student_data
    name = student_data["name"]

    # أضيف رسالة ترحيب من الروبوت تتضمن اسم الطالب
    chat_history.append(("bot", f"😄اهلا {name}"))

    # أضيف رسالة تحفيزية لبدء الأسئلة
    chat_history.append(("bot", f"✨هل انت مستعد يا {name} لبدا الاسئلة ؟ "))

    # أعيد رسم المحادثة لعرض رسائل الترحيب
    render_chat()


# في النهاية أستدعي دالة show_name
# وذلك لبدء تسلسل إدخال بيانات الطالب من أول خطوة
show_name()
```
